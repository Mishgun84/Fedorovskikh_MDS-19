## Уровень 0:

Посчитайте значение первой и второй производной функции:
<br><br>
<div> f(x) = x<sup>5</sup> + 4sin(2x) + cos(3x + 3) </div>

В точке x = 1.

Запишем первую производную:
<div> f'(x) = (x<sup>5</sup>)' + (4sin(2x))' + (cos(3x + 3))' = 5x<sup>4</sup> + 4cos(2x)&middot;2 - sin(3x + 3)&middot;3 = 5x<sup>4</sup> + 8cos(2x) - 3sin(3x + 3)</div>
<br>
И вторую производную:
<div> f"(x) = (5x<sup>4</sup>)' + (8cos(2x))' - (3sin(3x + 3))' = 20x<sup>3</sup> - 8sin(2x)&middot;2 - 3cos(3x + 3)&middot;3 = 20x<sup>3</sup> - 16sin(2x) - 9cos(3x + 3)</div>

Рассчитаем значения первой и второй производной в точке x = 1:

<div> f'(1) =  5&middot;1<sup>4</sup> + 8cos(2&middot;1) - 3sin(3&middot;1 + 3) = 5 + 8cos(2) -3sin(6) </div>
<div> f"(1) = 20&middot;1<sup>3</sup> - 16sin(2&middot;1) - 9cos(3&middot;1 + 3) = 20 - 16sin(2) - 9cos(6)</div>

```python
import numpy as np
import matplotlib.pyplot as plt

df1 = 5 + 8*np.cos(2) - 3*np.sin(6)
df2 = 20 - 16*np.sin(2) - 9*np.cos(6)
print('Значение первой производной в точке x=1 равно ',df1)
print('Значение второй производной в точке x=1 равно ',df2)
```

```python
# Найдем производные с помощью Python использую библиотеку sympy
from sympy import *
# создаем "символ" под названием x
x = Symbol('x')

# Определяем в переменную f нашу функцию
f = x**5 + 4*sin(2*x) + cos(3*x + 3)

# Находим первую производную
f_derivative = f.diff(x)
f_derivative
```

```python
# Находим вторую производную
f_2nd_derivative = f_derivative.diff(x)
f_2nd_derivative
```

```python
fd = lambdify(x, f_derivative)
f2d = lambdify(x, f_2nd_derivative)
print('Значение первой производной в точке x=1 равно ',fd(1))
print('Значение второй производной в точке x=1 равно ',f2d(1))
```

```python
# Производные и их значения в точке, найденные "вручную" совпадают с формулами и значениями, полученными с помощью 
# библиотеки sympy
```

## Уровень 1:

Постройте график первой и второй производной функции:
<br><br>
<div> f(x) = sin<sup>5</sup>(2x + 1) </div>

На отрезке [-5; 5].
    

Запишем первую производную:
<div> f'(x) = (sin<sup>5</sup>(2x + 1))' = 5sin<sup>4</sup>(2x + 1)&middot;(sin(2x+1))' = 10sin<sup>4</sup>(2x + 1)&middot;cos(2x + 1) </div>
<br>
<div>Можно применить тригонометрическую формулу синуса двойного угла и записать первую производную в виде:
 f'(x) = 5sin<sup>3</sup>(2x + 1)&middot;sin(4x + 2) но, тогда вторая производня будет иметь более сложный вид.</div>

<br>
Запишем вторую производную:
<div> f"(x) = 10((sin<sup>4</sup>(2x + 1))'&middot;cos(2x + 1) + sin<sup>4</sup>(2x + 1)&middot;(cos(2x + 1))') = 10(8sin<sup>3</sup>(2x + 1)&middot;cos<sup>2</sup>(2x + 1) - 2sin<sup>5</sup>(2x +1)) = 20(4sin<sup>3</sup>(2x + 1)&middot;cos<sup>2</sup>(2x + 1) - sin<sup>5</sup>(2x +1))</div>

```python
# Также найдем производные с помощью Python использую библиотеку sympy
from sympy import *
# создаем "символ" под названием x
x = Symbol('x')

# Определяем в переменную f нашу функцию
f = sin(2*x + 1)**5

# Находим первую производную
f_derivative = f.diff(x)
f_derivative
```

```python
# Находим вторую производную
f_2nd_derivative = f_derivative.diff(x)
f_2nd_derivative
```

```python
# Так же, производные, найденные "вручную" совпадают с формулами, полученными с помощью библиотеки sympy
```

```python
# Для построения графика зададим координаты по оси X. Думаю тысячи точек хватит для детализации.
X = np.linspace(-5, 5, num=1000, endpoint=True)

# И зададим выражения для координат по оси Y у первой и второй производной
Y1 = 10*(np.sin(2*X+1)**4)*np.cos(2*X+1)
Y2 = 20*(4*(np.sin(2*X+1)**3)*(np.cos(2*X+1)) - (np.sin(2*X+1)**5))
```

```python
# Построим графики производных
plt.figure(figsize=(10, 10))
plt.plot(X,Y1, color='green', label='f\'(x)')
plt.plot(X,Y2, color='red', label='f\"(x)')
plt.legend(fontsize=20)
plt.show()
```
